# show image grayscale & rgb

```python
#diplay the original and gray image
# Load the input image
image = cv2.imread('./istockphoto.jpg')
cv2.imshow('Original', image)
cv2.waitKey(0)

# Use the cvtColor() function to grayscale the image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
 
cv2.imshow('Grayscale', gray_image)
cv2.waitKey(0) 
 
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```

```python
#this displays gray image only 
# Use the second argument or (flag value) zero
# that specifies the image is to be read in grayscale mode
img = cv2.imread('./istockphoto.jpg', 0)
 
cv2.imshow('Grayscale Image', img)
cv2.waitKey(0)
 
# Window shown waits for any key pressing event
cv2.destroyAllWindows()
```

#  Median Filter

```python
# image path 
path = './istockphoto.jpg'

# using imread()  
img = cv2.imread(path)
dst = cv2.medianBlur(img,7)

cv2.imshow('image', np.hstack((img, dst)))
cv2.waitKey(0);
cv2.destroyAllWindows();
cv2.waitKey(1)
```

# Averaging Filter

```python
# image path 
path = './istockphoto.jpg'

# using imread()  
img = cv2.imread(path)

im1 = cv2.blur(img,(5,5))
im2 = cv2.boxFilter(img, -1, (2, 2), normalize=True)  

cv2.imshow('image', np.hstack((im1, im2)))
cv2.waitKey(0);
cv2.destroyAllWindows();
cv2.waitKey(1)
```

# Remove noise

```python
image = cv2.imread('./istockphoto.jpg',1) 

image_bw = cv2.imread('./istockphoto.jpg',0) 
```

```python
noiseless_image_bw = cv2.fastNlMeansDenoising(image_bw, None, 20, 7, 21) 

noiseless_image_colored = cv2.fastNlMeansDenoisingColored(image,None,20,20,7,21)
```

```python
titles = ['Image after removing the noise (colored)','Original Image(colored)', 'Image after removing the noise (grayscale)','Original Image (grayscale)']
images = [image,noiseless_image_colored, image_bw, noiseless_image_bw]
plt.figure(figsize=(13,5))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
plt.show()
```
